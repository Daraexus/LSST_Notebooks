## Coaddition of Images

```python
def Coaddition(path='coaddImages', tract=0, filter=None, visitList=None):
    
    visitStr = ''
    for visit in visitList : visitStr = visitStr +'^'+str(visit) 
    visitStr = visitStr[1:]
    print visitStr
    
    CoaddDir = DATADIR+'/output/coadd_r_dir/'+path
    
    if(os.path.exists(CoaddDir)==False):
        os.makedirs(CoaddDir)

    subprocess.call(["ln","-s",DATADIR+"/output", DATADIR+"/output/calexp/_parent"])
    SMTask = SkyMapper.MakeSkyMapTask
    res=SMTask.parseAndRun(args=(DATADIR+"/output/calexp --config  skyMap.active.decList=52.5,53.5  skyMap.active.raList=214.0,215.0 skyMap.active.patchInnerDimensions=1000,1000 --clobber-config --output "+DATADIR+"/output/coadd_r_dir/"+path).split())
    
        
    #Delimiting the zone
    stdout = sys.stdout
    sys.stdout = open(DATADIR+"/patches.txt", 'w')
    RPTask = PatchReporter.ReportPatchesTask
    res=RPTask.parseAndRun(args=(DATADIR+'/output/coadd_r_dir/'+path+' --config raDecRange=214.0,52.5,215.0,53.5 --id tract='+str(tract)+' patch=0,0 filter='+filter+' --clobber-config').split()) 
    sys.stdout = open(DATADIR+"/patches_r.txt", 'w')
    subprocess.call(["sed", "-e", 's/^/--id filter=r /', DATADIR+"/patches.txt"],stdout=sys.stdout)
    sys.stdout =stdout
    
    #Create the patches to be coadded
    TCTask = TempexpCoadder.MakeCoaddTempExpTask
    res=TCTask.parseAndRun(args=((DATADIR+"/output/calexp --output "+DATADIR+"/output/coadd_r_dir/"+path+" @"+DATADIR+"/patches_r.txt --selectId filter="+filter+" visit="+visitStr).split()))

    #Stack the patches
    ACTask = Assembler.AssembleCoaddTask
    res=ACTask.parseAndRun(args=((DATADIR+"/output/coadd_r_dir/"+path+" @"+DATADIR+"/patches_r.txt --selectId filter="+filter+" visit="+visitStr).split()))
```

# Subtraction:

```python
scienceFolder = "subtract+0"
templateFolder = "template"

SciencePath = DATADIR+"/output/coadd_r_dir/"+scienceFolder+"/deepCoadd/r/0/21,34.fits"
TemplatePath = DATADIR+"/output/coadd_r_dir/"+templateFolder+"/deepCoadd/r/0/21,34.fits"
OutputPath = DATADIR+"/subtracted"
```

```python
def SubtractImages(fwhm=None, templateImagePath=None, scienceImagePath=None):
    config = ipDiffim.ImagePsfMatchTask.ConfigClass()
    #Kernel name is Alard-Lupton by default
    config.kernel.name = "AL"
    subconfig = config.kernel.active
    #Fwhm will be calculated if none is provide
    defFwhm         = fwhm
    fwhmS = defFwhm
    fwhmT = defFwhm
    psfmatch = ipDiffim.ImagePsfMatchTask(config)

    #for image1, image2 in zip(Set1Files,Set2Files):
    #    print("Subtracting " + image1 + " " + image2)
    scienceImage = afwImage.ExposureF(scienceImagePath)
    templateImage = afwImage.ExposureF(templateImagePath)
    
    results  = psfmatch.subtractExposures(templateImage, scienceImage,
                            templateFwhmPix = fwhmT, scienceFwhmPix = fwhmS, doWarping=True)

    differenceExposure = results.subtractedExposure
    
    
    return differenceExposure
    
```
