These are the paths of the images for the subtraction. By definition, we are going to have two different images:

<b>Science Exposure:</b> This is the image that contains the variability of interest, in our particular case, that contains the variable/transient objects we would like to detect using the subtraction.
<b>Reference Exposure:</b> This is the template image that we are going to subtract from the Science Exposure. Usually this image corresponds to a capture with the same filter but in a diffrent moment of time.



```python
ScienceExposurePath = DATADIR+"/output/coadd_r_dir/subtract/deepCoadd/r/1/33,15.fits"
ReferenceExposurePath = DATADIR+"/output/coadd_r_dir/deepCoadd/r/1/33,15.fits"
SubtractionOutputPath = DATADIR+"/output/subtracted-inter"
```

<u>SubtractExposures</u> method call. We are accessing the <u>ImagePsfMatchTask</u> within <b>ip_diffim</b> in order to use this function. Another implementation is present in obs_cfht with the name of <u>ImageDifferenceTask</u>.
Config definition is handled beforehand. The kernel type for the PSF-matching will be obtained using the Alard-Lupton method (1998), hence the "AL" parameter.

Initial parameters declaration

```python
config = ipDiffim.ImagePsfMatchTask.ConfigClass()
config.kernel.name = "AL"
subconfig = config.kernel.active
defFwhm         = None
scienceFwhmPix = defFwhm
templateFwhmPix = defFwhm
psfmatch = ipDiffim.ImagePsfMatchTask(config)

psfmatch.kConfig.sizeCellX = 128
psfmatch.kConfig.sizeCellY = 128

ScienceExposure = afwImage.ExposureF(ScienceExposurePath)
ReferenceExposure = afwImage.ExposureF(ReferenceExposurePath)

#Uncomment to inverse the order
#ScienceExposureTemp = ScienceExposure
#ScienceExposure = ReferenceExposure
#ReferenceExposure = ScienceExposureTemp

convolveTemplate=True
doWarping = False
candidateList=None
frame = 1
```

## Step 1: Image Warping of input images

The method should validate the WCS of the ReferenceExposure and the ScienceExposure and in case doWarping is True, images are warped to adjust their WCS to the same reference frame

```python
if not psfmatch._validateWcs(ReferenceExposure, ScienceExposure):
            print "WCS were not validates. Attepting to warp the exposures"
            if doWarping:
                #self.log.info("Astrometrically registering template to science image")
                ReferenceExposure = psfmatch._warper.warpExposure(ScienceExposure.getWcs(),
                    ReferenceExposure, destBBox=ScienceExposure.getBBox(afwImage.PARENT))
            else:
                raise RuntimeError("Input images not registered")
```

Getting the FwhmPix for the ReferenceExposure

```python
 if templateFwhmPix is None:
            if not ReferenceExposure.hasPsf():
                print("No estimate of Psf FWHM for template image")
            else:
                templateFwhmPix = psfmatch.getFwhmPix(ReferenceExposure.getPsf())
print "templateFwhmPix = ",templateFwhmPix
```

Getting the FwhmPix for the ScienceExposure

```python
if scienceFwhmPix is None:
            if not ScienceExposure.hasPsf():
               print("No estimate of Psf FWHM for science image")
            else:
                scienceFwhmPix = psfmatch.getFwhmPix(ScienceExposure.getPsf())
print "scienceFwhmPix = ",scienceFwhmPix
```

Defining the size of the psf-matching kernel

```python
kernelSize = makeKernelBasisList(psfmatch.kConfig, templateFwhmPix, scienceFwhmPix)[0].getWidth()
```

This method generates a list of candidates for the PSF-matching

```python
candidateList = psfmatch.makeCandidateList(ReferenceExposure, ScienceExposure, kernelSize, candidateList)
```

In this part of the code, the subtraction is actually performed

```python
subtractedExposure = afwImage.ExposureF(ScienceExposure, True)



if convolveTemplate:
     subtractedMaskedImage  = subtractedExposure.getMaskedImage()
     subtractedMaskedImage -= results.matchedExposure.getMaskedImage()
        
     displayImage(subtractedMaskedImage.getImage(), frame=frame, title="Science-Matched",path="Science-Matched.png")
     frame+=1
    
        
     subtractedMaskedImage -= results.backgroundModel
     displayImage(subtractedMaskedImage.getImage(), frame=frame, title="Science-Matched-background", path="Science-Matched-background.png")
     frame+=1 
        
else:
     subtractedExposure.setMaskedImage(results.warpedExposure.getMaskedImage())
     subtractedMaskedImage  = subtractedExposure.getMaskedImage()
     subtractedMaskedImage -= results.matchedExposure.getMaskedImage()
     subtractedMaskedImage -= results.backgroundModel
        
     # Preserve polarity of differences
     subtractedMaskedImage *= -1
     #ds9.mtv(subtractedMaskedImage.getImage(), frame=frame, title="Polarity of differences preserved"); frame+=1
     #print "Polarity of differences preserved"
     #displayFigure("figure5.png")
        
     # Place back on native photometric scale
     subtractedMaskedImage /= results.psfMatchingKernel.computeImage(afwImage.ImageD(results.psfMatchingKernel.getDimensions()), False)
     #ds9.mtv(subtractedMaskedImage.getImage(), frame=frame, title="Back to native Photometric scale"); frame+=1
     #print "Back to native Photometric scale"
     #displayFigure("figure6.png")
```

```python
def displayFigure(path):
    i =  Image(path)
    display_png(i)
    
def displayImage(Image,frame=0,title="NoName",path="fig"):
    ds9.mtv(Image, frame=frame, title=title)
    ds9.ds9Cmd("zscale")
    ds9.ds9Cmd("zoom to fit")
    ds9.ds9Cmd("raise")
    ds9.ds9Cmd('saveimage '+path)
    print title
    displayFigure(path)

def showFrame(frame=0,title="NoTitle",path="fig"):
    ds9.ds9Cmd("zscale")
    ds9.ds9Cmd("zoom to fit")
    ds9.ds9Cmd("raise")
    ds9.ds9Cmd('saveimage '+path)
    print title
    displayFigure(path)
    
def displayImageWithSources(Image,frame=0,title="NoName",path="fig", sources=[]):
    ds9.mtv(Image, frame=frame, title=title)
    ds9.ds9Cmd("zscale")
    ds9.ds9Cmd("zoom to fit")
    ds9.ds9Cmd("raise")
    
    
    for source in sources:
        ds9.dot("+", source.getX()-Image.getX0()  , source.getY()-Image.getY0() , frame=frame, size = 25, ctype = ds9.RED)

    
    ds9.ds9Cmd('saveimage '+path)
    print title
    displayFigure(path)



```
