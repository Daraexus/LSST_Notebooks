```python
import lsst.afw.image as afwImage
import lsst.afw.display.ds9 as ds9
import os 
```

This is the same code used to test the remote usage of stack. The only addition is that the image is being saved to disk in order to display it later.

Note that, as ds9 is running on the local machine, the image will be stored locally and not on the server. This is going to be somehow addressed by this notebooks.

```python
DATADIR="/renoir_data_02/jpreyes/lsst_data/data/CFHTLS_TEST_SUB"

Path = DATADIR+"/subtracted+0/33,15.fits"

Image = afwImage.ExposureF(Path)

ds9.mtv(Image.getMaskedImage().getImage(), frame=1)
ds9.ds9Cmd('saveimage test.png')
```

Using this line we make a copy of the local stored image to the remote directory on the server side. This only work if we have configured previously the ssh public and private keys to avoid the need to use passwords confirmation.
