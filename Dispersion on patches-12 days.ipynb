# Dispersion on coadded images with independent jointcal on reference and science sets

```python
import numpy as np
import matplotlib.pyplot as plt
import json
%matplotlib inline

plt.figure( figsize=(20,5))

plt.subplot(1,2,1)

f = open('/datadec/cppm/jpreyes/CFHT_Jointcal_2/dispersion_ra_ind_12days.dat').read()
data_ra = json.loads(f)

plt.hist(data_ra, bins = 100) 
plt.xlabel("RA position difference in arcsec")
plt.ylabel("# of sources")


plt.subplot(1,2,2)

f = open('/datadec/cppm/jpreyes/CFHT_Jointcal_2/dispersion_dec_ind_12days.dat').read()
data_dec = json.loads(f)

plt.hist(data_dec, bins = 100) 
plt.xlabel("Dec position difference in arcsec")
plt.ylabel("# of sources")
#plt.ticklabel_format(style='sci', axis='x',scilimits=(0,0))

plt.show()
```

```python
dists_ind= np.sqrt(np.power(data_ra,2)+np.power(data_dec,2))

plt.hist(dists_ind, bins = 100) 
plt.xlabel("position difference in arcsec")
plt.ylabel("# of sources")

plt.show()
```

# Comparisons with match_raDec

```python

f = open('/datadec/cppm/jpreyes/CFHT_Jointcal_2/dispersion_ind_12days.dat').read()
data_ind_dist = json.loads(f)

f = open('/datadec/cppm/jpreyes/CFHT_Jointcal_2/dispersion_comp_12days.dat').read()
data_com_dist = json.loads(f)

f = open('/datadec/cppm/jpreyes/CFHT_Jointcal_2/dispersion_comp_par_12days.dat').read()
data_par_dist = json.loads(f)

f = open('/datadec/cppm/jpreyes/CFHT_Jointcal_2/dispersion_vanilla_12days.dat').read()
data_van_dist = json.loads(f)
```

```python
fig, ax0 = plt.subplots(ncols=1, figsize=(20,10))


ax0.hist(data_ind_dist, bins = 100, histtype="step", label="independent data with jointcal") 
ax0.text(0.25, 0.6,"Median scatter - %0.1f mas"%(np.median(np.asarray(data_ind_dist))), fontsize=20, verticalalignment='bottom', 
        transform=ax0.transAxes, color='blue',)
ax0.hist(data_com_dist, bins = 100, histtype="step", label="complete data with jointcal") 
ax0.text(0.25, 0.55,"Median scatter - %0.1f mas"%(np.median(np.asarray(data_com_dist))), fontsize=20, verticalalignment='bottom', 
        transform=ax0.transAxes, color='green',)


ax0.legend()
ax0.set_xlabel("position difference in mas")
ax0.set_ylabel("# of sources")

plt.show()
```

```python
fig, ax0 = plt.subplots(ncols=1, figsize=(20,10))

ax0.hist(data_van_dist, bins = 100, histtype="step", label="Data without jointcal") 
ax0.text(0.25, 0.6,"Median scatter - %0.1f mas"%(np.median(np.asarray(data_van_dist))), fontsize=20, verticalalignment='bottom', 
        transform=ax0.transAxes, color='blue',)
ax0.hist(data_ind_dist, bins = 100, histtype="step", label="independent data with jointcal") 
ax0.text(0.25, 0.55,"Median scatter - %0.1f mas"%(np.median(np.asarray(data_ind_dist))), fontsize=20, verticalalignment='bottom', 
        transform=ax0.transAxes, color='green',)


ax0.legend()
ax0.set_xlabel("position difference in mas")
ax0.set_ylabel("# of sources")
plt.show()
```
