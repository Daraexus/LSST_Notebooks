# Dispersion on coadded images with independent jointcal on reference and science sets

```python
import numpy as np
import matplotlib.pyplot as plt
import json
%matplotlib inline

plt.figure( figsize=(20,5))

plt.subplot(1,2,1)

f = open('/datadec/cppm/jpreyes/CFHT_Jointcal_2/dispersion_ra_ind_12days.dat').read()
data_ra = json.loads(f)

plt.hist(data_ra, bins = 100) 
plt.xlabel("RA position difference in arcsec")
plt.ylabel("# of sources")


plt.subplot(1,2,2)

f = open('/datadec/cppm/jpreyes/CFHT_Jointcal_2/dispersion_dec_ind_12days.dat').read()
data_dec = json.loads(f)

plt.hist(data_dec, bins = 100) 
plt.xlabel("Dec position difference in arcsec")
plt.ylabel("# of sources")
#plt.ticklabel_format(style='sci', axis='x',scilimits=(0,0))

plt.show()
```

```python
dists_ind= np.sqrt(np.power(data_ra,2)+np.power(data_dec,2))

plt.hist(dists_ind, bins = 100) 
plt.xlabel("position difference in arcsec")
plt.ylabel("# of sources")

plt.show()
```

# Dispersion on coadded images without Jointcal

```python
plt.figure( figsize=(20,5))

plt.subplot(1,2,1)

f = open('/datadec/cppm/jpreyes/CFHT_Jointcal_2/dispersion_ra_vanilla_12days.dat').read()
data_ra = json.loads(f)

plt.hist(data_ra, bins = 100) 
plt.xlabel("RA position difference in arcsec")
plt.ylabel("# of sources")


plt.subplot(1,2,2)

f = open('/datadec/cppm/jpreyes/CFHT_Jointcal_2/dispersion_dec_vanilla_12days.dat').read()
data_dec = json.loads(f)

plt.hist(data_dec, bins = 100) 
plt.xlabel("Dec position difference in arcsec")
plt.ylabel("# of sources")
#plt.ticklabel_format(style='sci', axis='x',scilimits=(0,0))

plt.show()
```

```python
plt.figure( figsize=(20,10))

plt.hist(dists_par, bins = 100, histtype="step", label="Coadd with jointcal parallelized")
plt.hist(dists_ind, bins=100, histtype ="step", label="Coadd with jointcal independent")
plt.hist(dists_comp, bins=100, histtype ="step", label="Coadd with jointcal complete")

plt.ylabel("# of objects")
plt.xlabel("radial distance (arcsec)")
plt.legend()
plt.show()
```
