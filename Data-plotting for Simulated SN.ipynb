```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

def get_arrays(sigmas):
    xs=[]
    ys=[]
    for sigma in sigmas:
        detected_sn = []
        other_detections = []
        parameters = []
        for row in sigma:

            r_data = [x for x in row.as_void()]
            sns = np.sum(r_data[1:-1])
            detected_sn.append(sns)
            other_detections.append(r_data[-1]-sns)
            parameters.append(r_data[0])
        xs.append(other_detections)
        ys.append(detected_sn)
    return xs, ys, parameters

def plot_family(xs, ys, sigmas_par, parameters, title="title", xmin=0, ymin=0):
    #fig = plt.figure(figsize=(10, 6))
    fig, ax = plt.subplots(figsize=(15,10))

    for x, y in zip(xs, ys):

        ax.scatter(x, y)

        for i, txt in enumerate(parameters):
            ax.annotate(txt, (x[i],y[i]))
    ax.legend(sigmas_par, title="Sigma of Injection")
    ax.set_xlim(xmin=xmin)
    ax.set_ylim(ymin=ymin)
    ax.set_title(title)
    ax.set_xlabel("Other DIASources")
    ax.set_ylabel("Injected Supernovae Found")
    
    plt.show()
```

```python
from astropy.table import Table
t3 = Table.read('t3sko-3.fits')
t5 = Table.read('t5sko-3.fits')
t10 = Table.read('t10sko-3.fits')
t15 = Table.read('t15sko-3.fits')
t20 = Table.read('t20sko-3.fits')

sigmas_par = [3,5,10,15,20] 
sigmas = [t3,t5,t10,t15,t20]

xs, ys, pars = get_arrays(sigmas)
plot_family(xs, ys, sigmas_par, pars, title="SKO")
```
