```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Sigmas = [0,5,10,15,20,25]

i_v = [25, 24,24,24,24,24]
i_t = [157375, 57413, 55989, 54009, 50791, 45707]

r_v = [28, 23,23,23,23,23]
r_t = [46382, 24683, 24335, 24018, 23646, 23027]

g_v = [13,10,10,10,10,10]
g_t = [33792,18197, 17917, 17605, 17270, 16789]
```

```python
t_t = [67806, 39381, 38656, 38059, 37372, 36555]
t_v = [28,27,27,27,26,26]
p_t = [32063, 33188, 32931, 32649, 32244, 31651]
p_v = [22,24,24,24,23,23]
```

```python
def anotate_plot(ax, Sigmas, r_v):
    for x, y in zip(Sigmas, r_v):
        ax.annotate(str(y), xy=(x,y))
```

```python
fig, ax = plt.subplots(figsize=(10,7))

plt.plot(Sigmas, r_v, label="filter r")
anotate_plot(ax, Sigmas, r_v)
plt.plot(Sigmas, i_v, label="filter i")
anotate_plot(ax, Sigmas, i_v)
plt.plot(Sigmas, g_v, label="filter g")
anotate_plot(ax, Sigmas, g_v)
plt.ylim(ymin=0, ymax=40)
ax.axhline(y=33, ls="dashed", c="PURPLE")
anotate_plot(ax, Sigmas, [33]*len(i_v))
ax.legend()
ax.set_xlabel("sigma factor")
ax.set_ylabel("# of SN found")
plt.show()

fig, ax = plt.subplots(figsize=(10,7))

plt.plot(Sigmas, r_t, label="filter r")
anotate_plot(ax, Sigmas, r_t)
plt.plot(Sigmas, i_t, label="filter i")
anotate_plot(ax, Sigmas, i_t)
plt.plot(Sigmas, g_t, label="filter g")
anotate_plot(ax, Sigmas, g_t)
plt.ylim(ymin=0)
ax.set_xlabel("sigma factor")
ax.set_ylabel("# of Candidates found")
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,7))
plt.plot(Sigmas, t_v, label="base")
anotate_plot(ax, Sigmas, t_v)
plt.plot(Sigmas, p_v, label="pos. proportion")
anotate_plot(ax, Sigmas, p_v)
plt.ylim(ymin=0, ymax=40)
ax.axhline(y=33, ls="dashed", c="PURPLE")
anotate_plot(ax, Sigmas, [33]*len(t_v))
ax.set_xlabel("sigma factor")
ax.set_ylabel("# of SN found")
ax.legend()
plt.show()

fig, ax = plt.subplots(figsize=(10,7))
plt.plot(Sigmas, t_t, label="base")
anotate_plot(ax, Sigmas, t_t)
plt.plot(Sigmas, p_t, label="pos. proportion")
anotate_plot(ax, Sigmas, p_t)
plt.ylim(ymin=0)
ax.set_xlabel("sigma factor")
ax.set_ylabel("# of Candidates found")
ax.legend()
plt.show()
```
